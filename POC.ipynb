```python
# Import Library.
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Read date from file.
df = pd.read_csv('jobs.csv')
print(df.columns)
df.head()
```

```python
# Feature Engineering.
df = df.iloc[0:1000, 0:8]
df.drop(['Job Salary', 'Job Experience Required'], axis=1, inplace=True)
df['Role Category'] = df['Role Category'].str.replace('/', ' ')
df['Key Skills'] = df['Key Skills'].str.replace('|', ' ')
df['Functional Area'] = df['Functional Area'].str.replace(',', ' ')
df['Industry'] = df['Industry'].str.replace(',', ' ')
df['Job Title'] = df['Job Title'].str.replace('/', ' ')
# print(df.columns)
df['complete string'] = df['Role Category'] + df['Key Skills'] + df['Functional Area'] + df['Industry'] + df['Job Title']
user_input = input('Search Your Job')
df['user input'] = user_input
df.tail(20)
```

```python
# Model Operations.
cv = CountVectorizer(lowercase=True, ngram_range=(1, 2))
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, 2))
tfidf_matrix = tfidf_vectorizer.fit_transform(df['complete string'])
user_matrix = tfidf_vectorizer.transform(df['user input'])

# Calc Cosine diff to check the distance parameter.
cosine_similarities = linear_kernel(user_matrix, tfidf_matrix)
df['Similarity'] = cosine_similarities[0]

# Plotting results.
new_data = df.sort_values(by='Similarity', ascending=False).iloc[:, 3:]
final_data_indexes = new_data.head().index.tolist()
df.iloc[final_data_indexes]
# df.iloc[168]
# print(new_data.iloc[0:1, :])
```
